```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Define the data
names = ["Kolbjrn", "Maja", "Henning", "Jacob", "Svein-Thore", "Mathilde", "Andreas"]
scores = {
    "Kolbjrn": [7, 7, 7, 7, 7, 7, 7],
    "Maja": [6, 6, 6, 6, 6, 7, 6],
    "Henning": [5, 5, 5, 5, 5, 6, 6],
    "Jacob": [5, 5, 4, 5, 5, 5, 4],
    "Svein-Thore": [4, 4, 3, 4, 4, 4, 3],
    "Mathilde": [4, 4, 3, 4, 4, 3, 3],
    "Andreas": [1, 1, 1, 1, 1, 1, 1]
}

# Prepare data for seaborn
data = []
for name in names:
    for score in scores[name]:
        data.append({"Person": name, "Score": score})

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Define variables
var_name = "Person"
df[var_name] = df[var_name].astype(str)

# Create a violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x=var_name, y="Score", data=df, inner="box", cut=1, color="tab:blue", linewidth=1.5, density_norm="count")

# Customize y-axis ticks
plt.gca().set_yticks([1, 2, 3, 4, 5, 6, 7])
plt.gca().set_yticklabels(['Lite plass', '', '', 'Middels plass', '', '', 'Masse plass'])

# Add labels and title
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.title("Ta plass, gi plass", fontsize=20)

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
