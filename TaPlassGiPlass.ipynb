```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
names = ["Kolbjørn", "Maja", "Henning", "Jacob", "Svein-Thore", "Mathilde", "Andreas"]
scores = {
    "Kolbjørn": [7, 7, 7, 7, 7, 7, 7],
    "Maja": [6, 6, 6, 6, 6, 7, 6],
    "Henning": [5, 5, 5, 5, 5, 6, 6],
    "Jacob": [5, 5, 4, 5, 5, 5, 4],
    "Svein-Thore": [4, 4, 3, 4, 4, 4, 3],
    "Mathilde": [4, 4, 3, 4, 4, 4, 3],
    "Andreas": [1, 1, 1, 1, 1, 1, 1]
}

# Calculate mean and standard deviation for each person
means = {name: np.mean(scores[name]) for name in names}
std_devs = {name: np.std(scores[name]) for name in names}

# Create a bar plot with error bars
fig, ax = plt.subplots()
ax.bar(means.keys(), means.values(), yerr=std_devs.values(), capsize=5)

# Add labels and title
ax.set_xlabel('Person')
ax.set_ylabel('Poeng')
ax.set_title('Ta Plass, Gi Plass. Høy verdi = masse plass.')

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
